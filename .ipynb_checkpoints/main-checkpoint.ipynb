```python
from mylib.lib import (
    load_dataset,
    grab_mean,
    grab_median,
    grab_std,
    create_bar,
    create_line
)

dataset = "https://raw.githubusercontent.com/fivethirtyeight/data/master/births/US_births_1994-2003_CDC_NCHS.csv"
```

```python
df = load_dataset(dataset)
assert df is not None
```

```python
df.describe()
```

```python
mean = grab_mean(df)
median = grab_median(df)
std = grab_std(df)

test = df.describe()
assert mean == test.iloc[1,1]
assert std == test.iloc[2,1]
assert median == test.iloc[5,1]
```
